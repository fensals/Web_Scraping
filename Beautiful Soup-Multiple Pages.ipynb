```python
#Importing all necessary libraries
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import numpy as np
from time import sleep
from random import randint
```

```python
#Initiating data storage by Creating empty lists to contain the data
Business = []
Phone_number = []
Address = []
Age_of_business = []
```

```python
#Creating a list containing containing numbers 1 to 4.
#This list will be iterated through for pages 1 to 4 of the website.
pages = np.arange(1,5)
```

```python
#Creating a loop to iterate through the 4 pages and extracting the HTML content.
for page in pages:
    page = requests.get("https://www.yellowpages.com/search?search_terms=plumber&geo_location_terms=AL&page=" + str(page))
    soup = BeautifulSoup(page.content, 'html.parser')
    results = soup.find(class_= "search-results organic")
    company = results.find_all('div', class_='result')
    #this will vary the amount of waiting time between requests for a number between 2-10 seconds
    sleep(randint(2,10)) 
    #Creating a loop to extract content from each page extracted from the previous loop.
    for comp in company:
        business_name = comp.find('a', class_='business-name')
        phone = comp.find('div', class_='phones phone primary')
        address = comp.find('div', class_='locality')
        years_in_business = comp.find('div', class_= 'number')
        if None in (business_name, phone, address, years_in_business): #skips rows with missing values
            continue
        #adding the selected contents into the empty lists
        Business.append(business_name.text)
        Phone_number.append(phone.text)
        Address.append(address.text)
        Age_of_business.append(years_in_business.text)
```

```python
#Combining the lists into a dataframe named YellowPages
YellowPages = pd.DataFrame({
'Business': Business,
'Phone_number': Phone_number,
'Address': Address,
'Age_of_business': Age_of_business,
})
```
